```python
# import libraries
import pandas as pd 
from plotly.offline import init_notebook_mode,iplot,download_plotlyjs,plot
from chart_studio import tools
import plotly.express as px
import plotly.graph_objs as go
import chart_studio
import numpy as np

init_notebook_mode  (connected = True)
```

```python
df = pd.read_csv(r"C:\Users\tolayemi\Documents\novel-corona-virus-2019-dataset\time_series_covid_19_recovered.csv")
df =df.drop(columns=['Province/State','Lat','Long'])
df.head()
```

```python
df = df[df['Country/Region'].isin(['France','US','Italy','Spain','Germany','China'])]

df =df.groupby('Country/Region').sum()
df
```

```python

#trace = go.Scatter(x=)

trace_one =go.Scatter(x=df.columns,
                     y= df.loc['China'],
                     name ='China',
                     line =dict(color='#092A54'),
                     opacity =0.8)
trace_two =go.Scatter(x=df.columns,
                     y= df.loc['US'],
                     name ='US',
                     line =dict(color='red'),
                     opacity =0.8)
trace_Three =go.Scatter(x=df.columns,
                     y= df.loc['Italy'],
                     name ='Italy',
                     line =dict(color='#7F7F7F'),
                     opacity =0.8)
trace_four =go.Scatter(x=df.columns,
                     y= df.loc['France'],
                     name ='France',
                     line =dict(color='cyan'),
                     opacity =0.8)
trace_five =go.Scatter(x=df.columns,
                     y= df.loc['Germany'],
                     name ='Germay',
                     line =dict(color='green'),
                     opacity =0.8)
data =[trace_one, trace_two,trace_Three,trace_four,trace_five]

layout = dict(title ='Recovered Covid-19 Patients by country ')
fig = dict(data=data, layout =layout)
iplot(fig,filename ='Recovered Covid-19 Patients')
```
